```sql
-- C. Acceso a Gold (Datamarts para consumo ejecutivo)
GRANT USE SCHEMA ON SCHEMA `bank_dev`.`gold` TO `DA's`;
GRANT SELECT ON TABLE `bank_dev`.`gold`.`fraud_transaction_datamart` TO `DA's`;
```

```sql
-- GESTIÓN DE SEGURIDAD POR TABLA (Granularidad)

-- Los Analistas no deben ver los datos de 'bronze' (datos sensibles/crudos)
-- Si intentan hacer un SELECT, Unity Catalog denegará el acceso automáticamente
REVOKE SELECT ON SCHEMA `bank_dev`.`bronze` FROM `DA's`;

-- Aseguramos acceso preciso a las dimensiones en Silver
GRANT SELECT ON TABLE `bank_dev`.`silver`.`transactions` TO `DA's`;
GRANT SELECT ON TABLE `bank_dev`.`silver`.`dim_users` TO `DA's`;
GRANT SELECT ON TABLE `bank_dev`.`silver`.`dim_cards` TO `DA's`;
```
